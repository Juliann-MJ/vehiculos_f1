```python
# ======================================================================
# AVERAGE OF NULL VALUES IN THE COLUMN: : "model_year".
# ======================================================================
promedio_model_year = df['model_year'].isnull().sum() / len(df["model_year"]) * 100
print(f'Promedio de valores nulos en la columna "model_year": {promedio_model_year:.2f}%')

# ======================================================================
# REMOVE MISSING VALUES OF THE COLUMN: "model_year".
# ======================================================================
df = df.dropna(subset=['model_year'])

# ======================================================================
# CONVERT THE VALUES TO INTEGERS OF THE COLUMN: "model_year".
# ======================================================================
df['model_year'] = df['model_year'].astype(int)
```

* Se realizo el calculo promedio de los valores nulos de la columna "model_year" arrojandonos un resultado de: 7.02%, lo que representa una pequeña fraccion, lo que nos lleva a la conclucion de eliminar por completo esos valores nulos.

##### **_5.12 Column: paint_color._**

```python
# ======================================================================
# SHOW COLUMN "paint_color".
# ======================================================================
df['paint_color']
```

```python
# ======================================================================
# SHOW UNIQUE VALUES OF THE COLUMN "paint_color".
# ======================================================================
df['paint_color'].unique()
```

```python
# ======================================================================
# FILL MISSING VALUES WITH: UNKNOWN OF THE COLUMN "paint_color".
# ======================================================================
df['paint_color'] = df['paint_color'].fillna('unknown')
print(f'Valores nulos en la columna "paint_color": {df['paint_color'].isna().sum()}')
```

```python
# ======================================================================
# IMPUTE MISSING VALUES WITH 0 OF THE COLUMN "is_4wd".
# ======================================================================
df['is_4wd'] = df['is_4wd'].fillna(0)
print(f'Valores nulos en la columna "is_4wd": {df['is_4wd'].isna().sum()}')

# ======================================================================
# CONVERT THE VALUES TO INTEGER OF THE COLUMN "is_4wd".
# ======================================================================
df['is_4wd'] = df['is_4wd'].astype(int)
```

* Podemos verificar que los valores de la columna "is_4wd" manejan datos de tipo float, lo que se requiere manejar datos de tipo int, dicho procedimiento fue realizado con exito. tambien podemos notar presencia de valores ausentes, por lo que se dictamino que esos valores ausentes fueran inpotados con valor 0, esto con el fin de mantener la esencia del dataframe sin perder este tipo de valores, por el tratamiento de estos datos ahora quedaron en formato correcto.
